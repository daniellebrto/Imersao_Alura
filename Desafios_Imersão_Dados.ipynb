### Desafio 01: Investigar por que a classe tratamento é tão desbalanceada?

```python
import pandas as pd
url = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true' 
dados = pd.read_csv (url, compression = 'zip')
dados ['tratamento'].value_counts()
```

Considerando que o controle representa um cenário utilizado para comparação dos resultados obtidos com a droga concluímos que não é necessária a existência do mesmo número de resultados de controle e de drogas, uma vez que um controle pode ser utilzado como comparação de diversos resultados de experimentos com drogas.

###Desafio 07: Deixar os gráficos bonitões. (Matplotlib.pyplot)

```python
import matplotlib.pyplot as plt
dados['tempo'].value_counts().plot.bar()
plt.title('Tempo de duração das doses')
ax = plt.plot(data='tempo')
plt.ylabel('Qtdade de Experimentos')
plt.xlabel('Tempo (em horas)')
plt.autumn()
plt.show()
```
